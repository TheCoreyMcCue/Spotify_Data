```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('./Desktop/spotify2/tracks.csv')
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df = df_tracks.sort_values('popularity', ascending = False).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis=1).corr(method='pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt ='.1g', vmin = -1, vmax = 1, center = 0, cmap ='inferno', linewidths = 1, linecolor = 'black')
heatmap.set_title('Correlation Heatmap')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize = (10, 6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness to Energy Correlation")
```

```python
plt.figure(figsize = (10, 6))
sns.regplot(data = sample_df, y = "popularity", x = "loudness", color = "b").set(title = "Popularity to Loudness Correlation")
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Number of Songs by Year")
```

```python
total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Duration of Songs by Year")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax).set(title = "Duration of Songs by Year")
plt.xticks(rotation = 90)
```
